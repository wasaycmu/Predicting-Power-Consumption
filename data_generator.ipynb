## We merge gendered information

```python
import pandas as pd

# Assuming 'data' is your DataFrame loaded with census data
# List of columns to retain and modify
columns_to_keep = [
    'state', 'total_pop', 'households'
]

# Define age brackets to merge
age_brackets = [
    ('under_5', 'male_under_5', 'female_under_5'),
    ('5_to_9', 'male_5_to_9', 'female_5_to_9'),
    ('10_to_14', 'male_10_to_14', 'female_10_to_14'),
    ('15_to_17', 'male_15_to_17', 'female_15_to_17'),
    ('18_to_19', 'male_18_to_19', 'female_18_to_19'),
    ('20', 'male_20', 'female_20'),
    ('21', 'male_21', 'female_21'),
    ('22_to_24', 'male_22_to_24', 'female_22_to_24'),
    ('25_to_29', 'male_25_to_29', 'female_25_to_29'),
    ('30_to_34', 'male_30_to_34', 'female_30_to_34'),
    ('35_to_39', 'male_35_to_39', 'female_35_to_39'),
    ('40_to_44', 'male_40_to_44', 'female_40_to_44'),
    ('45_to_49', 'male_45_to_49', 'female_45_to_49'),
    ('50_to_54', 'male_50_to_54', 'female_50_to_54'),
    ('55_to_59', 'male_55_to_59', 'female_55_to_59'),
    ('60_to_61', 'male_60_61', 'female_60_to_61'),
    ('62_to_64', 'male_62_64', 'female_62_to_64'),
    ('65_to_66', 'male_65_to_66', 'female_65_to_66'),
    ('67_to_69', 'male_67_to_69', 'female_67_to_69'),
    ('70_to_74', 'male_70_to_74', 'female_70_to_74'),
    ('75_to_79', 'male_75_to_79', 'female_75_to_79'),
    ('80_to_84', 'male_80_to_84', 'female_80_to_84'),
    ('85_and_over', 'male_85_and_over', 'female_85_and_over')
]

# Adding age bracket columns to the list of columns to keep and modifying data to merge male and female counts
for bracket, male_col, female_col in age_brackets:
    data[f'aged_{bracket}'] = data[male_col] + data[female_col]
    columns_to_keep.append(f'aged_{bracket}')

# Filter the DataFrame to keep only the specified columns
filtered_data = data[columns_to_keep]

# Filter by state
filtered_data = filtered_data[filtered_data['state'].isin(['CA', 'TX', 'AZ', 'TN'])]

# Print the modified DataFrame to verify the result
print("\nColumns in the modified dataset:")
print(filtered_data.columns.tolist())

# Save to CSV
filtered_data.to_csv('filtered_census_dataset.csv', index=False)

print("Filtered dataset saved.")
```

## Rollup by state

```python
import pandas as pd

# Assuming filtered_data is already loaded and contains the state-filtered data
# Perform aggregation: sum all numeric columns for each state
state_aggregated_data = filtered_data.groupby('state').sum()

# Optionally, you can reset the index to make 'state' a column again
state_aggregated_data.reset_index(inplace=True)

# Print the aggregated DataFrame to verify the result
print("\nAggregated Data by State:")
print(state_aggregated_data)

# Save to CSV
state_aggregated_data.to_csv('state_aggregated_census_dataset.csv', index=False)

print("State aggregated dataset saved.")
```

## Rebucketing into our age groups

```python
# Rebucket age groups
state_aggregated_data['num_age_0_to_18'] = (
    state_aggregated_data['aged_under_5'] + 
    state_aggregated_data['aged_5_to_9'] + 
    state_aggregated_data['aged_10_to_14'] + 
    state_aggregated_data['aged_15_to_17']
)

state_aggregated_data['num_age_19_to_40'] = (
    state_aggregated_data['aged_18_to_19'] + 
    state_aggregated_data['aged_20'] + 
    state_aggregated_data['aged_21'] + 
    state_aggregated_data['aged_22_to_24'] + 
    state_aggregated_data['aged_25_to_29'] + 
    state_aggregated_data['aged_30_to_34'] + 
    state_aggregated_data['aged_35_to_39']
)

state_aggregated_data['num_age_41_to_62'] = (
    state_aggregated_data['aged_40_to_44'] + 
    state_aggregated_data['aged_45_to_49'] + 
    state_aggregated_data['aged_50_to_54'] + 
    state_aggregated_data['aged_55_to_59'] + 
    state_aggregated_data['aged_60_to_61']
)

state_aggregated_data['num_age_63_to_all'] = (
    state_aggregated_data['aged_62_to_64'] + 
    state_aggregated_data['aged_65_to_66'] + 
    state_aggregated_data['aged_67_to_69'] + 
    state_aggregated_data['aged_70_to_74'] + 
    state_aggregated_data['aged_75_to_79'] + 
    state_aggregated_data['aged_80_to_84'] + 
    state_aggregated_data['aged_85_and_over']
)

# Drop the original age columns they are no longer needed
state_aggregated_data.drop(columns=[
     'aged_under_5', 'aged_5_to_9', 'aged_10_to_14', 'aged_15_to_17', 
     'aged_18_to_19', 'aged_20', 'aged_21', 'aged_22_to_24', 'aged_25_to_29', 
     'aged_30_to_34', 'aged_35_to_39', 'aged_40_to_44', 'aged_45_to_49', 
     'aged_50_to_54', 'aged_55_to_59', 'aged_60_to_61', 'aged_62_to_64', 
     'aged_65_to_66', 'aged_67_to_69', 'aged_70_to_74', 'aged_75_to_79', 
     'aged_80_to_84', 'aged_85_and_over'
 ], inplace=True)

# Print the updated DataFrame to verify the result
print("\nUpdated Aggregated Data by State with New Age Buckets:")
print(state_aggregated_data)

# Save to CSV
state_aggregated_data.to_csv('state_aggregated_census_dataset_rebucketed.csv', index=False)

print("State aggregated dataset with new age buckets saved.")
```

## Normalize per household
create a new dataset called data_population where all the features are normalized per household,  will divide each demographic feature by the number of households in each state. This allows to see the average number of individuals per household and the distribution across the different age brackets on a per-household basis.

```python
import pandas as pd

# Copy the original aggregated dataset
data_population = state_aggregated_data.copy()

# List of columns to normalize by the number of households
columns_to_normalize = [
    'total_pop', 'num_age_0_to_18', 'num_age_19_to_40', 
    'num_age_41_to_62', 'num_age_63_to_all'
]

# Normalize each column by the number of households
for column in columns_to_normalize:
    data_population[column] = data_population[column] / data_population['households']

# Drop the 'households' column after normalization
data_population.drop('households', axis=1, inplace=True)

# Print the updated DataFrame to verify the result
print("\nData per Household for Each State (without households column):")
print(data_population)

# Save to CSV
data_population.to_csv('state_population_per_household.csv', index=False)

print("Dataset with population per household saved without 'households' column.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_house_size = pd.read_csv('./raw_house_size.csv')

# Group data by 'State' and calculate average house size
average_house_size = data_house_size.groupby('state')['average_size'].mean().sort_values()

# Plotting the data
plt.figure(figsize=(10, 8))
average_house_size.plot(kind='bar')
plt.title('Average House Size by State')
plt.xlabel('State')
plt.ylabel('Average House Size (sq ft)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import norm

# Cluster Plot
plt.figure(figsize=(10, 4))
states = data_house_size['state'].unique()
average_sizes = data_house_size.groupby('state')['average_size'].mean()
plt.scatter(states, average_sizes, alpha=0.6)
plt.xticks(rotation=90)
plt.title('Cluster Plot of Average House Size by State')
plt.xlabel('State')
plt.ylabel('Average House Size (sq ft)')

# Normal Distribution of House Sizes
plt.figure(figsize=(10, 6))
sns.distplot(data_house_size['average_size'], fit=norm, kde=True)
plt.title('Normal Distribution of House Sizes')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Frequency')

plt.show()
```

```python
for city_info in cities:
    state_code = city_info['state_iso']  # This line sets the state_code variable
    area_average_m2 = average_house_size.loc[city_info['state']] * feet_to_meters_squared
    # Retrieve the average number of people per household for the current state
    household_size = state_age_group_data[state_code]['total_pop']
    
    for i in range(1, 1001):  # Generate 100 users per city
        # Initialize user features that don't change monthly
        user_base = {
            'username': str(i).zfill(6),
            'password': 'password',
            'failed_attempts': 0,
            'fk_referral': 1,
            'fk_user': 1,
            'pk_dwelling': 1,
            'pk_resident': 1,
            'pk_user': 1,
            'ts_last_login': '',
            'ts_updated': '',
            'hash': '',
            'ts_created': '',
            'ts_deleted': '',
            'num_age_0_to_18': np.random.poisson(state_age_group_data[state_code]['num_age_0_to_18'] * household_size),
            'num_age_19_to_40': np.random.poisson(state_age_group_data[state_code]['num_age_19_to_40'] * household_size),
            'num_age_41_to_62': np.random.poisson(state_age_group_data[state_code]['num_age_41_to_62'] * household_size),
            'num_age_63_to_all': np.random.poisson(state_age_group_data[state_code]['num_age_63_to_all'] * household_size),
            'when_moved_in': (datetime(1965, 1, 1) + timedelta(days=random.randint(0, 21549))).strftime('%Y-%m-%d'),
            'when_moved_out': '',
            'compass_facing': random.choice(compass_facing),
            'type': np.random.choice(probabilities_dwelling_types.index, p=probabilities_dwelling_types.values),
            'level': random.randint(1, 8),
            'energy_grading': random.randint(15, 96),
            'material_walls': np.random.choice(probabilities_wall_materials.index, p=probabilities_wall_materials.values),
            'roof_material': np.random.choice(probabilities_roof_materials.index, p=probabilities_roof_materials.values),
            'material_windows': random.choice(windows_materials),
            'has_double_glacing': random.randint(0, 1),
            'has_heating': 1,
            'has_hot_water': 1,
            'has_air_conditioning': 1,
            'has_dishwasher': random.randint(0, 1),
            'has_washer': random.randint(0, 1),
            'has_drier': random.randint(0, 1),
            'has_fridge': 1,
            'has_pool': random.randint(0, 1),
            'num_entertainment': random.randint(1, 3),
            'kwh_generation': 0,
            'street': ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=random.randint(5, 10))),
            'city': city_info['city'],
            'state': city_info['state_iso'],
            'country': city_info['country'],
            'post_code': random.choice(list(city_info['post_codes'])),
            'area': np.random.normal(area_average_m2, area_std_dev_m2)
        }

        # Generate bills for each month
        for j in range(1, 13):
            month_days = calendar.monthrange(2023, j)[1]
            kwh = random.randint(300, 1700)
            price_pretax = round(random.randint(13, 22) * kwh * 0.01, 2)
            price_posttax = round((1 + salestax) * price_pretax, 2)

            bill = user_base.copy()
            bill.update({
                'date_start': datetime(2023, j, 1).strftime('%Y-%m-01'),
                'date_end': datetime(2023, j, month_days).strftime('%Y-%m-%d'),
                'kwh': kwh,
                'currency': 'USD',
                'price_pretax': price_pretax,
                'price_posttax': price_posttax,
                'pk_bill': '',
                'tariff': ''
            })
            user_data_list.append(bill)
```

```python
# Convert list to DataFrame
synthetic_data_df = pd.DataFrame(user_data_list)

# Save to CSV
synthetic_data_df.to_csv(output_file_name, index=False)

print(f"Synthetic data generated and saved to '{output_file_name}'.")
```
